```python
bad_bins = []
for i, w in enumerate(gbm_tte.bins.widths):

    if w < 5e-2:
        bad_bins.append(i)


edges = [gbm_tte.bins.starts[0]]

for i, b in enumerate(gbm_tte.bins):

    if i not in bad_bins:
        edges.append(b.stop)

starts = edges[:-1]
stops = edges[1:]


gbm_tte.create_time_bins(starts, stops, method="custom")
```
